According to the bussiness requirements, we are here to determine whether there is any difference in average TAT among the different Laboratories

Significance: 5% <br>
alpha = 0.05 <br>
alpha/2 = 0.025 <br>

<b>Null Hyppothesis:</b> there is no difference in the average Turn Around Time (TAT) of reports of the laboratories on their preferred list <br>
<b>Alternate Hypothesis:</b> there is a difference in average TAT atleast in one Laboratory.

<b> Performing ANOVA (Analysis of Variance) on the independent numerical columns

```python
import scipy.stats as stats
```

```python
t_static, p_value = stats.f_oneway(df["Laboratory 1"], df["Laboratory 2"], df["Laboratory 3"], df["Laboratory 4"])
```

```python
print("t_static: ", t_static)
print("p_value: ", p_value)
```

```python
alpha = 0.05
if p_value<=alpha:
    print("We reject Null Hypothesis, there is a sifnificance difference between TAT")
else:
    print("We fail to reject Null Hypothesis")
```
